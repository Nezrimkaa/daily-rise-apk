# 📱 Daily Rise - Сборка APK

**Инструкция:**
1. Загрузи ZIP-архив с папкой `android-app` когда появится окно
2. Жди 20-30 минут
3. APK скачается автоматически

```python
# === ВСЁ В ОДНОЙ ЯЧЕЙКЕ - ПРОСТО ЗАПУСТИ ===

# 1. Установка зависимостей
!pip install buildozer cython==0.29.36
!sudo apt-get update
!sudo apt-get install -y git zip unzip openjdk-11-jdk autoconf libtool pkg-config zlib1g-dev cmake libffi-dev libssl-dev libncursesw5-dev libncurses-dev

# 2. Загрузка ZIP
from google.colab import files
import os, zipfile, glob

os.makedirs('/content/apk', exist_ok=True)
os.chdir('/content/apk')

print("\n" + "="*50)
print("📦 ЗАГРУЗИ ZIP С ПАПКОЙ android-app")
print("="*50)
uploaded = files.upload()

# 3. Распаковка
for f in uploaded.keys():
    if f.endswith('.zip'):
        zipfile.ZipFile(f).extractall('.')
        print(f"✅ Распаковано: {f}")

os.chdir('/content/apk/android-app')
print(f"\n📁 Папка: {os.getcwd()}")

# 4. Сборка
print("\n" + "="*50)
print("🔨 НАЧАЛАСЬ СБОРКА (20-30 минут)")
print("="*50)
!buildozer android debug

# 5. Скачивание
print("\n" + "="*50)
print("📥 СКАЧИВАНИЕ APK...")
print("="*50)
apk = glob.glob('bin/*.apk')
if apk:
    files.download(apk[0])
    print("✅ APK скачан!")
else:
    print("❌ APK не найден. Проверь логи выше.")
```
