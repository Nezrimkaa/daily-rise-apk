**Инструкция:**

2. Загрузи **ZIP-архив** с папкой `android-app`
3. Жди **20-30 минут**
4. APK скачается автоматически

```python
# ═══════════════════════════════════════════════════════════════
# 📱 DAILY RISE - APK BUILDER
# ═══════════════════════════════════════════════════════════════
# Просто нажми ▶ и загрузи ZIP когда появится окно
# ═══════════════════════════════════════════════════════════════

# 1. Установка зависимостей
print("📦 Установка зависимостей...")
!pip install buildozer cython==0.29.36 -q
!sudo apt-get update -qq
!sudo apt-get install -y -qq git zip unzip openjdk-11-jdk autoconf libtool pkg-config zlib1g-dev cmake libffi-dev libssl-dev libncursesw5-dev libncurses-dev
print("✅ Зависимости установлены!")

# 2. Загрузка ZIP
from google.colab import files
import os, zipfile, glob

os.makedirs('/content/apk', exist_ok=True)
os.chdir('/content/apk')

print("\n" + "="*50)
print("📦 ЗАГРУЗИ ZIP С ПАПКОЙ android-app")
print("="*50)
uploaded = files.upload()

# 3. Распаковка
print("\n📂 Распаковка...")
for f in uploaded.keys():
    if f.endswith('.zip'):
        zipfile.ZipFile(f).extractall('.')
        print(f"✅ Распаковано: {f}")

os.chdir('/content/apk/android-app')
print(f"\n📁 Папка: {os.getcwd()}")
!ls -la

# 4. Сборка
print("\n" + "="*50)
print("🔨 СБОРКА APK (20-30 минут)")
print("НЕ ЗАКРЫВАЙ ВКЛАДКУ!")
print("="*50)
!buildozer android debug

# 5. Скачивание
print("\n" + "="*50)
print("📥 СКАЧИВАНИЕ APK...")
print("="*50)
apk = glob.glob('bin/*.apk')
if apk:
    files.download(apk[0])
    print("✅ APK скачан!")
    print(f"📍 Файл: {apk[0]}")
else:
    print("❌ APK не найден. Проверь логи выше.")
```
